```python
import requests
from bs4 import BeautifulSoup

class Daft:
    _ENDPOINT = "https://www.daft.ie/property-for-rent/ireland/houses"
    _HEADER = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Content-Type": "application/json",
        "brand": "daft",
        "platform": "web",
    }

    def get_html_content(self):
        response = requests.get(
            self._ENDPOINT,
            headers=self._HEADER
        )
        
        if response.status_code == 200:
            return response.content
        else:
            print(f"Failed to retrieve HTML content. Status code: {response.status_code}")
            return None

    def extract_links(self, html_content):
        links = []
        soup = BeautifulSoup(html_content, 'html.parser')
        results = soup.find_all('li', {'data-testid': 'results'})
        for result in results:
            link = result.find('li', {'data-testid': 'result-4577972'}).find('a')['href']
            links.append(link)
        return links

# Initialize Daft instance
daft_instance = Daft()

# Get HTML content
html_content = daft_instance.get_html_content()

if html_content:
    # Extract links
    links = daft_instance.extract_links(html_content)
    
    # Print all links
    for link in links:
        print(link)
else:
    print("No HTML content retrieved")
```
